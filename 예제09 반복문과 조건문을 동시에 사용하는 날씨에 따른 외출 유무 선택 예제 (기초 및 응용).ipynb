```python
#(1)
#시간 모듈 import
import time

#비나 눈이 오는지 질문
rain=input("지금 비나 눈이 오나요?(y/n):")

#비가 올 경우 우산의 유무를 질문
if rain=="y" :
    umbrella=input("그렇다면 우산이 있나요?(y/n):" )
    #우산이 없을경우(if) 비가 그칠때 까지 기다림(whlie)
    if umbrella=="n":
        while rain=="y":
            print("비가 그치기를 기다립시다")
            time.sleep(1)#비가 그치기 기다리는 동안 1초가 지남
            rain=input("지금도 비나 눈이 오나요?(y/n):")

#비가 오지 않는다면 해가 쨍쨍한지를 물어봄
else :
    sun=input("오늘 해가 쨍쨍하나요?(y/n):")
    #해가 쨍쨍하다면 썬블락의 유무를 물어봄
    if sun=="y":
        sun_block=input("썬블락이 집에 있나요?(y/n):")
        #썬블락이 없을경우(if) 해가 조금 사라지기를 기다림(while)
        if sun_block=="n":
            while sun=="y":
                print("해가 조금 사라질때를 기다립시다")
                time.sleep(1)#해가 사라지길 기다리는 동안 1초가 지남
                sun=input("지금도 해가 쨍쨍한가요?(y/n):")

#옷을 입기 위해 날씨가 따뜻함의 여부를 물어봄
print("준비가 됬군요. 옷을 입어 봅시다")

#따뜻할 경우 얇은 옷을 입게함
hot=input("오늘 날씨가 덥나요?(y/n):")
if hot=="y" :
    print("얇게 입고 나가세요")
    
#날씨가 덥지 않을 경우(else) 추운지를 물어보고 추울 경우(if) 두꺼운 옷을 입게함
else :
    cold=input("그렇다면 춥나요?(y/n):")
    if cold=="y":
        print("따뜻하게 입고 나가세요")
    #날씨가 덥지도 춥지도 않을경우 적당한 옷을 입게함.
    else :
        print("적당하게 입고 나가세요")

#외출
print("외출합시다")
```

```python
#(2)
##[순서도 그리기-눈 내리는 날 외출하기]##



#import예약어를 통해 시간을 기다리는 모듈 time을 불러온다.
#nuber변수를 설정한다. 뒤에서 기다리는 것의 반복횟수 제한용도
import time
number=0

#현재 날씨를 묻고 이를 날씨를 알리는 변수에 넣는다.
isSnow=input("눈이 오는가?(y/n):")

#눈이 온다면, 우산을 가지고 있는지 묻는다.
if isSnow=="y":
    haveUmbrella=input("우산을 가지고 있는가?(y/n):")

    #눈이 오지만, 우산을 가지고 있다면, 외출한다.
    if haveUmbrella=="y":
        print("외출한다.")

    #눈이 오지만 우산을 가지고 있지 않는다면 잠시 기다린다.    
    else :              
        while number<5 and isSnow=="y":
            print("잠시 기다린다.")
            number=number+1
             #time모듈을 이용해 1분간 기다린 후 다시 묻는데,이를 while예약어로 반복.
            time.sleep(60)         
            isSnow=input("눈이 오는가?(y/n):")
         #이때 number변수를 만들어 5번까지만 질문을 반복하도록 한다
        if number>=5:
            print("눈이 그칠 기미가 없으므로 택시를 탄다.")
        else:
            print("외출한다.")

            
#눈이 오지 않는다면, 외출한다.
else:
    print("외출한다.")
```

```python
#(4)
#시간 모듈을 가져온다
import time

#택시 변수를 초기화한다.
taxi=0

#외출 여부를 물어본다.
outing=input("외출하시겠습니까?(y/n):")

#외출시 날씨를 물어본다.
if outing=="y":
    
    #비가 오는지 물어봅니다.
    rain=input("지금 비가 옵니까?(y/n):")

    #비가 온다면 우산의 유무를 물어본다.
    if rain=="y":
        umbrella=input("우산을 가지고 있나요?(y/n):")
        #우산이 없다면 날씨가 좋아질 때까지 기다릴지 택시를 탈지 선택한다.
        if umbrella=="n":
            taxi=input("그렇다면 택시를 타고 외출하시겠습니까?(y/n):")
            #택시를 탄다면 외출준비를 한다. 타지 않는다면 비가 그치기를 기다린다.
            if taxi=="n":
                while rain=="y":
                    print("날씨가 좋아질 때까지 기다립시다.")

                    #기다리는 동안 1초가 지난다
                    time.sleep(1)

                    #날씨가 좋아졌는지 물어본다.
                    rain=input("지금도 비가 오나요?(y/n):")

    #비가 오지 않는다면 햇빛이 강한지 물어본다.
    else :
        sun=input("지금 햇빛이 강한가요?(y/n):")

        #햇빛이 강하다면 자외선 차단제나 양산이 있는지 물어본다.
        if sun=="y":
            sun_block=input("자외선 차단제를 가지고 있나요?(y/n):")

            #자외선 차단제나 양산이 없다면 해가 약해지는 것을
            #기다릴지 택시를 타고 나갈지를 물어본다.
            if sun_block=="n":
                taxi=input("그렇다면 택시를 타고 외출하시겠습니까?(y/n):")

                #택시를 탄다면 외출준비를 한다. 타지 않는다면 해가 약해지기를 기다린다.
                if taxi=="n":
                    while sun=="y":
                        print("해가 약해질 때까지 기다립시다.")

                        #기다리는 동안 1초가 지난다
                        time.sleep(1)

                        #날씨가 좋아졌는지 물어본다.
                        sun=input("지금도 햇빛이 강하나요(y/n):")

    #외출을 하기로 했으니 옷을 입기위해 바깥 온도를 물어본다.
    print("외출을 위해 이제 옷을 골라봅시다.")
    temperature=input("현제 온도는 어떤가요?(hot, soso, cold):")

    #날씨에 따라 옷을 골라준다.
    if temperature=="hot":
        print("시원한 옷을 입으세요.")
    elif temperature=="cold":
        print("따뜻한 옷을 입으세요.")
    else :
        print("적당한 두께의 옷을 입으세요.")

    #옷을 입는 동안 기다린다. 택시를 부르기로 했다면 택시를 호출한다.
    time.sleep(1)
    if taxi=="y":
        print("출발이 얼마남지 않았습니다. 택시를 호출합니다.")

    #사람을 만나러 밖에 나가는 것인지 물어본다.
    meet=input("오늘 밖에서 사람을 만나나요?(y/n):")

    #사람을 만난다면 그 사람이 누구인지 물어본다.
    if meet=="y":
        partner=input("만나는 사람이 누구인가요?:")

        #그 사람에게 출발 메세지를 보낼지 물어본다.
        send_massage_partner=input("그 사람에게 출발 메세지를 보낼까요?(y/n):")

        #메세지를 보낼경우 메세지를 보낸다.
        if send_massage_partner=="y":
            time.sleep(1)
            print(partner+"에게 출발메세지를 보냈습니다.")

    #택시를 불렀을 경우 택시의 도착을 알려준다.
    if taxi=="y":
        time.sleep(2)
        print("택시가 도착했습니다.")

    print("외출합시다.")

#외출하지 않을경우 앱을 종료한다.
else :
    print("알겠습니다. 앱을 종료합니다.")
```
