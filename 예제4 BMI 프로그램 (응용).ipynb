```python
import sys
yes = ["응", "엉", "그래", "ㅇ", "ㅇㅇ", "d", "예", "네", "yes", "y", "Y"]             # 사용자 답변용 리스트. yes or no에 해당하는 예상 답변을 리스트에 기입하였음.
no = ["아니", "싫어", "ㄴ", "ㄴㄴ", "s", "아뇨", "no", "n", "N"]
group = ["WHO BMI", "아시아태평양지역 BMI"]                 #결과값 표시에 사용할 리스트
repeat = 1  #반복실행용 플래그
print ("안녕하세요? BMI 계산 프로그램입니다.")

while repeat == 1:
    choice = input("설명을 들으시겠어요? 아니면 계산을 하시겠어요?\n=>") # 이분법적으로 질문하여 사용자의 답변에 '설명' 혹은 '계산'이 포함되도록 유도.
    if choice.count("설명") >= 1:                                 # "설명해줘" 등의 대답 예시를 고려, [설명]이라는 단어가 포함되었을 경우 설명을 선택한 것으로 간주하도록 함.
        print ("BMI(Body Mass Index)는 미국의 보험 업계에서 고객의 신체 정보를 기록하던 것에서 유래했습니다.\n\
성별 등의 정보는 고려하지 않고 쉽게 측정할 수 있는 체중과 신장만을 이용하기 때문에 비만도 측정용으로 간단하게 사용할 수 있습니다.\n\
하지만 BMI는 실제 체지방률을 구하지 않기 때문에 상대적 척도로만 활용하는 것을 권장합니다.\n\
예를 들어, 사람의 근육 세포는 같은 부피의 지방에 비해 약 1.15배 무겁다고 합니다.\n\
따라서 같은 키의 보디빌더가 배불뚝이 부장님보다 BMI가 높게 나올 수도 있는 겁니다.")
        while 1:
            choice = input("계산을 해 보시겠어요?\n=>") #choice 변수는 매번 사용자 입력 시 일회성으로 사용하는 변수.
            if choice in yes:   # 위에 작성된 yes 리스트에 포함될 경우 yes로 간주
                print ("알겠습니다. 계산을 시작해 봅시다.")
                repeat = 0      # while이 두 개 겹쳐 사용되므로 한 번에 다음으로 넘어갈 수 있도록 플래그를 사용하였음.
                break
            elif choice in no:
                print ("아쉽네요. 잘 가요!")
                sys.exit()      # 프로그램 종료
            else:
                print ("무슨 말인지 모르겠어요!")
    elif choice.count("계산") >=1:
        print ("알겠습니다. 계산을 시작해 봅시다.")
        break
    else:
        print ("무슨 말인지 모르겠어요!")
        

while 1:
    while 1:
        standard = int(input("WHO 기준 BMI를 사용하시려면 1, 아시아태평양지역 기준 BMI를 사용하시려면 2를 입력하세요.\n=>")) #BMI 기준 선택
        if standard in [1,2]:
            print (group[standard-1],"선택.") # 개인 편의를 위해 standard를 1~2로 하였으나 리스트는 0부터 시작이므로 1을 빼서 사용
            break
        else:
            print ("다시 입력해 주세요.")
    height = float(input("키를 입력해 주세요. (단위 : cm)\n=>").replace("cm",''))      #키와 몸무게에 단위까지 입력하는 경우를 대비하여 단위를 걸러내도록 함.
    weight = float(input("몸무게를 입력해 주세요. (단위 : kg)\n=>").replace("kg",''))
    print ("키 :", height, "\n몸무게 :",weight)
    BMI = weight/(height/100)**2        #BMI 계산식
    print ("BMI 지수 =",round(BMI,2))     #소숫점 둘째 자리까지만 표시하도록 함
    print ("(이 수치는 최대 소수점 아래 두 번째 자리까지 표시됩니다)")
    
    if standard == 1:                   #뒤에서 결과를 프린트하기 위해 지정한 BMI 기준에 따라 결과를 결정함
        if BMI < 18.5:
            result = ("저체중입니다.")
        elif BMI < 25:
            result = ("정상입니다.")
        elif BMI < 30:
            result = ("과체중입니다.")
        elif BMI < 35:
            result = ("1단계 비만입니다.")
        elif BMI < 40:
            result = ("2단계 비만입니다.")
        else:
            result = ("고도비만입니다.")
    if standard == 2:
        if BMI < 18.5:
            result = ("저체중입니다.")
        elif BMI < 23:
            result = ("정상입니다.")
        elif BMI < 25:
            result = ("과체중입니다.")
        elif BMI < 30:
            result = ("비만입니다.")
        else:
            result = ("고도비만입니다.")
    print (group[standard-1], "기준에 따르면 당신은", result)
        
    while 1:
        choice = input("계속 하시겠어요? \n=>")
        if choice in yes:
            print ("좋아요! 다시 시작할게요!") #계산 재개
            break
        elif choice in no:
            print ("아쉽네요. 안녕, 잘 가요!") #프로그램 종료
            sys.exit()
        else:
            print ("무슨 말인지 모르겠어요!")
```
